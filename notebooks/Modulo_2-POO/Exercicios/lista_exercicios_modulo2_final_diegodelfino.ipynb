Crie uma classe Bola cujos atributos são cor e raio. Crie um método que imprime a cor da bola. Crie um método para calcular a área dessa bola. Crie um método para calcular o volume da bola. Crie um objeto dessa classe e calcule a área e o volume, imprimindo ambos em seguida.

Obs.:

Área da esfera = 4* 3.14* r* r;

Volume da esfera = (4/3)* 3.14* r* r* r

```python
PI = 3.14

class Bola:
    '''
    Classe que cria uma bola
    '''
    def __init__(self, cor, raio):
        '''
        Construtor cria a cor e o raio da bola.
        
        Parâmetros
        ----------
        cor : str
            cor da bola.
        raio : float
            raio da bola.
        '''
        self.cor = cor
        self.raio = raio
        
        # a partir do objeto criado com os métodos abaixo
        self.area = self.calcular_area()
        self.volume = self.calcular_volume()
    
    # Não é necessário docstring para os métodos mágicos __repr__ e __srt__
    def __repr__(self):
        return f'cor: {self.cor} \nraio: {self.raio}'
    
    def imprimir_cor(self):
        '''
        Imprime a cor da bola (self).
        '''
        print(f'A cor da bola é {self.cor}')
    
    def calcular_area(self):
        '''
        Calcula a área de uma circuferência.
        '''
        area = 4*PI*self.raio*self.raio
        return area
        
    def calcular_volume(self):
        '''
        Calcula o volume de uma esfera.
        '''
        esfera = (4/3)*PI*self.raio*self.raio*self.raio
        return esfera

### VALIDANDO O MÉTODO CONSTRUTOR e MÉTODO MÁGICO REPR
### Criando as bolas

bola = Bola(
    cor = 'verde',
    raio = 4)

bola1 = Bola(
    cor = 'vermelha',
    raio = 5.6)

# objeto bola criado

bola

# objeto bola1 criado

bola1

### VALIDANDO O MÉTODO imprimir_cor
### Imprime a cor da bola

# validando o método de imprimir a cor da bola

bola.imprimir_cor()

### VALIDANDO O MÉTODO calcular_area
### Cria o objeto área

# validando o método área

bola.area

### VALIDANDO O MÉTODO calcular_volume
### Cria o objeto volume

# validando o método volume

bola.volume

## Printando a área e volume
print(f'A área da bola é {bola.area:.5}m2')
print(f'O volume da bola é {bola.volume:.5}m3')
```

Crie uma classe Funcionario cujos atributos são nome e e-mail. Guarde as horas trabalhadas em um dicionário cujas chaves são o mês em questão e, em outro dicionário, guarde o salário por hora relativo ao mês em questão. Crie um método que retorna o salário mensal do funcionário.

```python
class Funcionario:
    '''
    Classe Funcionário
    '''
    
    def __init__(self, nome, email):
        '''
        Cria um funcionário
        
        Parâmetros
        ----------
        nome : str
            nome do funcionário
        email : str
            email do funcionário
        '''
        self.nome = nome
        self.email = email
        self.horas_trabalhadas = {}
        self.salario_hora = {}
        
    # Não é necessário docstring para os métodos mágicos __repr__ e __srt__
    def __repr__(self):
        return f'nome: {self.nome} \nsalário: {self.email}'
    
    def lancar_horas_mes(self, mes, quantidade_horas):
        '''
        Lança a quantidade de horas do mês
        
        Parâmetros
        ----------
        mês : str
            mês vigente, exemplo: "07/2021"
        quantidade_horas : float
            quantidade de horas trabalhadas no mês vigente
        '''
        self.horas_trabalhadas[mes] = quantidade_horas
        
    def definir_salario_hora_mes(self, mes, valor_salario_hora):
        '''
        Define o valor salário hora no mês vigente
        
        Parâmetros
        ----------
        mes : str
            mês vigente, exemplo: "07/2021"
        valor_salario_hora: float
            valor em hora do salário no mês vigente
        '''
        self.salario_hora[mes] = valor_salario_hora
        
    def calcular_salario(self, mes):
        '''
        Calcula o salário no mês vigente
        
        Parâmetros
        ----------
        mes : str
            mês vigente, exemplo: "07/2021"
            
        Retorno
        -------
        out : float
            Salário no mês vigente
            Retorna 0 se não trabalhou no mês
            ou se o mês não está cadastrado.
        '''
        if (mes in self.salario_hora and
           mes in self.horas_trabalhadas):
            calculo_salario = self.salario_hora[mes] * self.horas_trabalhadas[mes]
            return f'{calculo_salario:.7}'
        print ('Salário ou horas não cadastrado no mês solicitado')
        
diego = Funcionario(
    nome = 'Diego',
    email = 'dzionn@gmail.com')

print(diego)
print(diego.nome)

diego.definir_salario_hora_mes('07/2021', 68.18)
diego.lancar_horas_mes('07/2021', 220)

print(diego.calcular_salario('07/2021'))

print(diego.calcular_salario('06/2021'))
```

Enunciado

Crie uma classe Televisor cujos atributos são:

a. fabricante;

b. modelo;

c. canal atual;

d. lista de canais; e

e. volume.

Faça métodos para aumentar/diminuir volume, trocar o canal e sintonizar um novo canal, que adiciona um novo canal à lista de canais (somente se esse canal não estiver nessa lista). No atributo lista de canais, devem estar armazenados todos os canais já sintonizados dessa TV.

Obs.: O volume não pode ser menor que zero e maior que cem; só se pode trocar para um canal que já esteja na lista de canais.

```python
class Televisor:
    '''
    Classe Televisor
    '''
    def __init__(self, fabricante, modelo, canal_atual, volume):
        '''
        Cria um televisor
        fabricante : str
            Fabricante do televisor
        modelo : str
            Modelo do televisor
        canal_atual : int
            Canal de configuração inicial
        volume : int
            Volume atual do televisor
        '''
        self.fabricante = fabricante
        self.modelo = modelo
        self.volume = volume
        self.canal = canal_atual
        
        self.lista_canais = []
        if (canal_atual not in self.lista_canais):
            self.lista_canais.append(self.canal)
            
    # Não é necessário docstring para os métodos mágicos __repr__ e __srt__
    def __repr__(self):
        return f'fabricante: {self.fabricante} modelo: {self.modelo}'
        
    def alterar_volume (self, volume):
        '''
        Aumentar ou diminui o volume.
        
        Parâmetro
        ---------
        volume : int
            Número do volume
        Condição: O volume deve ser configurado entre 0 e 100
        '''
        if volume >= 0 and volume <= 100:
            self.volume = volume
            return (f'O volume do televisor {self.modelo} agora está {volume}.')
        else:
            return ('O volume deve ser entre 0 e 100')
            
    def sintonizar_canal(self, canal):
        '''
        Sintoniza os canais e o acrescenta a lista de canais
        
        Parâmetros
        ---------
        canal : int
            Canal a ser sintonizado
        '''
        if (canal not in self.lista_canais):    
            self.lista_canais.append(canal)
            
    def trocar_canal(self, novo_canal):
        '''
        Troca de canal e o acrescenta a lista de canais
        
        Parâmetros
        ----------
        novo_canal : int
            Substitui o canal atual e, se não estiver sintonizado, será acrescentado a lista
        '''
        self.canal = novo_canal
        if (novo_canal not in self.lista_canais):
            self.lista_canais.append(novo_canal)
            
# instanciando o objeto

Mac = Televisor(
    fabricante = 'Apple',
    modelo = 'MacBook Air',
    canal_atual= 4,
    volume = 1)

print(Mac)

# validando o método alterar_volume

print(Mac.alterar_volume(99))

# validando o método sintonizar_canal

print(f'A lista de canais é: {Mac.lista_canais}')
Mac.sintonizar_canal(10)
print(f'Após sintonizado, a lista de canais é: {Mac.lista_canais}')

# validando o atributo lista_canais

print(Mac.lista_canais)

# validando o método troca de canal

print(f'O canal era {Mac.canal}.')
Mac.trocar_canal(6)
print(f'O canal foi trocado para {Mac.canal}.')
print(f'A lista atual de canais é {Mac.lista_canais}.')

# validando o método alterar_volume

print(Mac.alterar_volume(5))
print(Mac.alterar_volume(-5))
print(Mac.alterar_volume(105))
```

## Questão #7

Enunciado
Crie uma modelagem de classes para uma agenda capaz de armazenar contatos. Através dessa agenda é possível incluir, remover, buscar e listar contatos já cadastrados.

```python
class Agenda:
    # necessário dicionário externo ao método construtor, pois
    # senão, ele ira subscrever o contato cada vez que um novo objeto for criado
    contatos = {}
    
    def __init__(self, contato, telefone):
        '''
        Método construtor
        
        Parâmetro
        ---------
        contato : str
            Nome do contato
        telefone : str
            Telefone do contato
        '''
        # definindo os atributos do método construtor
        self.contato = contato
        self.telefone = telefone
        
        # define um dicionário e insere o contato no dicionário
        Agenda.contatos.update({self:[self.contato,self.telefone]})
        
        # método __repr__ para representar o objeto
    def __repr__(self):
        return f'Contato {self.contato}, Telefone: {self.telefone}'
        
        # método para incluir o contato na agenda
        # que irá chamar o construtor para criar o objeto com os parâmetros/atributos da classe
        
    def incluir_contato(nome, telefone):
        '''
        Método que chama o construtor e cria o contato com os parâmetros/atributos da classe
        
        Parâmetros
        ----------
        nome : str
            Nome do contato
        telefone : int
            Telefone do contato
        '''
        if nome not in Agenda.contatos:
            return Agenda(nome, telefone)
        
    @staticmethod
    def remover_contato(nome):
        '''
        Método que chama o construtor e remove o contato no dicionário de contatos
        
        Parâmetro
        ---------
        nome : str
            Nome do contato
        '''
        for nome_existente in Agenda.contatos:
            if nome == Agenda.contatos[nome_existente][0]:
                del Agenda.contatos[nome_existente]
                print('Contato foi removido da agenda')
            else:
                print('Contato não existe')
            break
            
    # @staticmethod significa: quando esse método é chamado, não passamos uma instância d
    # a classe para ele (como fazemos normalmente com os métodos). 
    # Isso significa que você pode colocar uma função dentro de uma classe, 
    # mas não pode acessar a instância dessa classe (isso é útil quando o método não usa a instância).
    
    @staticmethod
    def buscar_contato(nome):
        '''
        Busca o contato na lista de contatos
        
        Parâmetro
        ---------
        nome : str
            Nome do contato
        '''
        for nome_existente in Agenda.contatos:
            if nome == Agenda.contatos[nome_existente][0]:
                print (f'Nome: {Agenda.contatos[nome_existente][0]}, Telefone: {Agenda.contatos[nome_existente][1]}')
    
    @staticmethod
    def listar_contatos():
        '''
        Listar todos os contatos cadastrados
        '''
        for nome_existente in Agenda.contatos:
            print (f'Nome: {Agenda.contatos[nome_existente][0]}, Telefone: {Agenda.contatos[nome_existente][1]}')

# validando o método incluir_contato
Agenda.incluir_contato('Evellyn', 11982838888)
Agenda.incluir_contato('Diego', 11982839001)
Agenda.incluir_contato('Wladimir', 55982839001)
Agenda.incluir_contato('Arthur', 88982839001)
Agenda.incluir_contato('Evellyn', 21982838888)


# validando o método lista_contatos
Agenda.listar_contatos()

# validando o método buscar_contato
Agenda.buscar_contato('Evellyn')

# validando o método remover_contato
Agenda.remover_contato('Evellyn')
Agenda.remover_contato('Diego')
Agenda.remover_contato('Fádua')
Agenda.remover_contato('Diego')
Agenda.listar_contatos()
```

```python
class Cliente:
    # dicionário de grupo de clientes que garanta a referencia para o objeto criado
    grupo_clientes = {}
    def __init__(self, nome, idade, email):
        '''
        Método construtor
        
        Parâmetros
        ----------
        nome : str
            Nome do cliente
        idade : int
            Idade do cliente
        email : str
            e-mail do cliente
        '''
        
        # atributos do método
        self.nome = nome
        self.idade = idade
        self.email = email
        # realizando adição dos objetos criados no dicionário
        # referenciando o e-mail como chave de identificação do objeto
        Cliente.grupo_clientes.update({self.email:[self.nome, self.idade, self.email]})
        
    def __repr__(self):
        return f'Nome: {self.nome}\nIdade: {self.idade}\nE-mail: {self.email}\n'

class Sistema():
    '''
    Uma classe simples para controlar a classe Clientes.
    Não será utilizado o método construtor, pois não serão atribuitos objetos a essa classe.
    '''
    def adicionar_cliente(nome, idade, email):
        '''
        Método para chamar os atributos da classe Cliente.
        '''
        return Cliente(nome, idade, email)
    
    @staticmethod
    def modificar_cliente(identificador, nome, idade, email):
        '''
        Método estático para editar informações na base de dados do cadastro do cliente.
        
        Parâmetros
        ----------
        identificador : obj
            Objeto criado na classe Clientes que será editado
        nome : str
            Nome do cliente
        idade : int
            Idade do cliente
        email : int
            e-mail do cliente
        '''
        Cliente.grupo_clientes.update({identificador:[nome, idade, email]})

    @staticmethod
    def imprime_clientes():
        '''
        Método estático para apresentar lista de clientes
        '''
        print(f'\nTotal de clientes cadastrados:', len(Cliente.grupo_clientes), '\n')
        for elementos in Cliente.grupo_clientes:
            print(f'Nome: {Cliente.grupo_clientes[elementos][0]}')
            print(f'Idade: {Cliente.grupo_clientes[elementos][1]}')
            print(f'e-mail: {Cliente.grupo_clientes[elementos][2]}\n')

opcao_menu = ''
print('Escolha uma das opções do menu')

# def valida_inteiro(idade):
#         if (not isinstance(idade, int) or idade<0):
#             try:
#                 idade = int(idade)
#             except:
#                 raise ValueError('A idade deve ser inteiro.')     
        
while opcao_menu != '4':
    opcao_menu = input('1. Adicionar cliente\n2. Modificar cadastro\n3. Mostrar clientes\n4. Sair\n\n')

    if opcao_menu == '1':
        nome = input('\n1.1-Insira o nome do cliente: ')
        idade = int(input('1.2-Insira a idade do cliente: '))
        email = input('1.3-Insira o e-mail do cliente: ')
        identificador = email

    # no dicionário ele percorre todas as chaves, assim não há necessidade de indicar a posição
        if identificador not in Cliente.grupo_clientes:
            Sistema.adicionar_cliente(nome, idade, email)
            print('Cliente adicionado com sucesso.\n')
            print(f'Adicionado a chave {Cliente.grupo_clientes[identificador]}\n')
        else:
            print('Cliente já possui cadastro.\n')

    #Modificar cadastro
    elif opcao_menu == '2':
        identificador = input('\n2.1-Insira o e-mail do cadastro:')
        if identificador in Cliente.grupo_clientes:
            nome = input('2.2-Insira o nome a ser alterado: ')
            idade = int(input('2.3-Insira a idade a ser alterada: '))
            email = identificador
            print('2.4-e-mail não pode ser modificado. Para alterar o e-mail faça novo cadastro.')
            Sistema.modificar_cliente(identificador, nome, idade, email)
            print('\nNome e/ou idade alterado(s) com sucesso.\n')
            print(f'\nModificado o cadastro para: {Cliente.grupo_clientes[identificador]}\n')
        else:   
            print('\ne-mail não localizado no cadastro.')

    elif opcao_menu == '3':
        Sistema.imprime_clientes()

    # if = se, elif = senão se, else = senão
    elif opcao_menu == '4':
        print('\nAté logo. Volte sempre!\n')
        break
    else:
        print('Ocorreu um erro\n')
```
