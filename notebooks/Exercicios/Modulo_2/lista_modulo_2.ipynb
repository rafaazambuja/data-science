## Questão 1 
    Crie uma classe Bola cujos atributos são cor e raio. Crie um método que imprime a cor da bola. Crie um método para calcular a área dessa bola. Crie um método para calcular o volume da bola. Crie um objeto dessa classe e calcule a área e o volume, imprimindo ambos em seguida. 
    Obs.:

    Área da esfera = 4*3.14*r*r;

    Volume da esfera = 4*3.14*r*r*r/3

````python
class Bola:
    def __init__(self, raio, cor):
        '''
        Construtor
        
        Parâmetros
        ------------
        raio : str
            raio da bola.
        cor : str
            cor da bola.        
        '''
        self.raio = raio
        self.cor = cor
        
    def calcula_area(self):
        '''
        Calcula a área da bola
        
        Retorno
        ---------
        out : int
            area da bola
        '''
        area = 4 * 3.14 * self.raio * self.raio 
        return area

    def calcula_volume(self):
        '''
        Calcula o volume da bola
        
        Retorno
        ---------
        out : int
            volume da bola
        '''
        volume = (4 / 3) * 3.14 * self.raio * self.raio * self.raio
        return volume
    
    def imprime_cor(self):
        '''
        Imprime a cor da bola
        '''
        print(f'A cor da bola é {self.cor}')
        
    def __repr__(self):
        return f" Raio: {self.raio} - Cor: {self.cor}"
````

````python
bola1 = Bola(4, 'preto')

print(f'A área da bola é {round(bola1.calcula_area(),2)}\n')
print(f'O volume da bola é {round(bola1.calcula_volume(),2)}\n')
bola1.imprime_cor()
````

## Questão 2 
    Crie uma classe Retângulo cujos atributos são lado_a e lado_b. Crie um método para calcular a área desse retângulo. Crie um objeto dessa classe e calcule a área e a imprima em seguida.

````python
class Retangulo:
    def __init__(self, lado_a, lado_b):
        '''
        Construtor
        
        Parâmetros
        ------------
        lado_a : int
            lado A do retângulo.
        lado_b : int
            lado B do retângulo
        '''
        self.lado_a = lado_a
        self.lado_b = lado_b
        
    def calcula_area(self):
        '''
        Calcula a área do retângulo
        
        Retorno
        -------
        out : int
            área do retângulo.
        '''
        return self.lado_a * self.lado_b
    
    def __repr__(self):
        return f" Lado A: {self.lado_a} - Lado B: {lado_b}"
````

## Questão 3
    Crie uma classe Funcionario cujos atributos são nome e e-mail. Guarde as horas trabalhadas em um dicionário cujas chaves são o mês em questão e, em outro dicionário, guarde o salário por hora relativo ao mês em questão. Crie um método que retorna o salário mensal do funcionário.

````python
class Funcionario:
    def __init__(self, nome, email):
        '''
        Contrutor
        
        Parâmetros
        -----------
        nome : str
            nome do funcionário
        email : str
            email do funcionário
        '''
        
        self.nome = nome
        self.email = email
        self.horas_trabalhadas = {}
        self.salario_hora = {}
        
    def __repr__(self):
        return f"Nome: {self.nome} - E-mail: {self.email}"
        
    def lancamento_horas(self, mes, horas):
        '''
        Lança a quantidade de horas do mês

        Parâmetros
        ------------
        mes : str
            mês referente as horas trabalhadas
        horas : float
            quantidade de horas trabalhadas
        '''
        self.horas_trabalhadas[mes] = horas

    def lancamento_valor_hora(self, mes, valor_hora):
        '''
        Lança o valor/hora do mês

        Parâmetros
        -----------
        mes : str
            mês referente ao valor/hora
        valor_hora : float
            valor em reais para 1 hora de trabalho
        '''
        self.salario_hora[mes] = valor_hora

    def calcula_salario(self, mes):
        '''
        Calcula o salário do funcionário no mês

        Parâmetros
        -----------
        mes : str
            mês que deseja consultar o salario

        Retorno
        -----------
        out : float
            Salário do mês consultado
        '''

        if mes in self.horas_trabalhadas:
            if mes in self.salario_hora:
                salario = self.horas_trabalhadas[mes] * self.salario_hora[mes]
                return salario
            else:
                print(f"Salário/hora do mês {mes} não foi cadastrado")
        else:
            print(f"Horas trabalhadas no mês {mes} não foram cadastradas")
````

````python
rafael = Funcionario('Rafael', 'rafabuja@gmail.com')

rafael.lancamento_horas(3, 35.4)
rafael.lancamento_valor_hora(3, 120)

rafael.calcula_salario(3)
````

## Questão 4
    Crie uma classe Televisor cujos atributos são:

    a. fabricante;

    b. modelo;

    c. canal atual;

    d. lista de canais; e

    e. volume.

    Faça métodos para aumentar/diminuir volume, trocar o canal e sintonizar um novo canal, que adiciona um novo canal à lista de canais (somente se esse canal não estiver nessa lista). No atributo lista de canais, devem estar armazenados todos os canais já sintonizados dessa TV.

    Obs.: O volume não pode ser menor que zero e maior que cem; só se pode trocar para um canal que já esteja na lista de canais.

````python
class Televisor:
    def __init__(self, fabricante, modelo, canal, volume):
        '''
        Contrutor
        
        Parâmetros
        -----------
        fabricante : str
            fabricante do televisor
        modelo : str
            modelo do televisor
        canal : int
            canal atual do televisor
        volume : int
            volume atual do televisor
        '''
        self.fabricante = fabricante
        self.modelo = modelo
        self.canal = canal
        self.volume = volume
        
        self.lista_canais = [canal]
    
    def __repr__(self):
        return f"Fabricante: {self.fabricante} - Modelo: {self.modelo}"
        
    def altera_volume(self, volume):
        '''
        Alterar o volume do Televisor
        
        Parâmetros
        -----------
        volume : int
            Novo volume do televisor
        '''
        if volume < 0 and volume > 100:
            raise ValueError("O volume tem que estar entre 0 e 100")
        else:
            self.volume = volume
    
    def troca_canal(self, canal):
        '''
        Alterar o canal do Televisor
        
        Parâmetros
        -----------
        canal : int
            Novo canal do televisor
        '''
        self.canal = canal
        
        if canal not in self.lista_canais:
            self.lista_canais.append(canal)
            
    def sintoniza_canal(self, canal):
        '''
        Sintoniza um canal do Televisor
        
        Parâmetros
        -----------
        canal : int
            Canal a ser sintonizado
        '''      
        if canal not in self.lista_canais:
            self.lista_canais.append(canal)
````

````python
tv = Televisor('AOC', '1', 12, 3)

tv.altera_volume(99)
tv.troca_canal(6)

print(tv.volume)

print(tv.canal)

print(tv.lista_canais)
````

## Questão 7
    Crie uma modelagem de classes para uma agenda capaz de armazenar contatos. Através dessa agenda é possível incluir, remover, buscar e listar contatos já cadastrados.

````python
class Agenda:
    def __init__(self):
        self.contatos = {}
    
    def incluir_contato(self, nome, telefone):
        '''
        Adiciona novos contatos a sua agenda
        
        Parâmetro
        ----------
        nome : str
            nome do contato
        telefone : str
            telefone do contato
        '''
        # O telefone por ser unico foi adicionado como chave do dicionário
        if telefone in self.contatos.keys():
            return f"Este numero de telefone já está cadastrado como {nome}"
        
        self.contatos.update({telefone:nome})
        
        
    def remover_contato(self, nome):
        '''
        Remove contatos a sua agenda

        Parâmetro
        ----------
        nome : str
            nome do contato que deseja remover
        '''
        for key, value in agenda.contatos.items():
            if nome == value:
                # problema: vai remover sempre o primeiro contato encontrado com o nome
                # eu não posso continuar no for depois de remover o contato (quebra o loop)
                del self.contatos[key]
                break
            
    def buscar_contato(self, nome):
        '''
        Imprime o contato bucado na agenda

        Parâmetro
        ----------
        nome : str
            nome do contato que deseja buscar
        '''
        for key, value in agenda.contatos.items():
            if nome == value:
                print(f"{value} : {key}")
                
    def listar_contatos(self):
        '''
        Imprime todos os contatos da agenda
        '''
        for key, value in agenda.contatos.items():
            print(f"{value} : {key}")
````

````python
agenda = Agenda()

agenda.incluir_contato('Rafael', '988054633')
agenda.incluir_contato('Mãe', '988864646')
agenda.incluir_contato('Teste', '000000000')

agenda.remover_contato('Teste')

agenda.buscar_contato('Mãe')
print()
agenda.listar_contatos()
````

````python
class Sistema:
    cadastro = {}
    def __init__(self):
        pass
    def cadastrar_cliente(self, nome, idade, email):
        '''
        Criar um objeto cliente e salvar no atributo cadastro
        
        Parâmetros
        -----------
        nome : str
            Nome do Cliente
        idade : int
            Idade do Cliente
        email : str
            E-mail do Cliente
        '''
        cliente = Cliente(nome, idade, email)
        
        # Não recomentado utilizar o nome como chave para acessar Cliente
        # Está sendo utilizado apenas em ambiente controlado
        sistema.cadastro.update({nome:cliente})
    
    def editar_cliente(self, key, nome=None, idade=None, email=None):
        '''
        Editar o objeto cliente
        
        Parâmetros
        -----------
        key : str
            Identificador do Cliente no dicionário (nome)
        nome : str
            Novo nome do cliente
        idade : int
            Nova idade do cliente
        email : str
            Novo e-mail do cliente
        ''' 
        # Editar apenas dado selecionado no programa (incompleto ainda)
        if idade is not None:
            sistema.cadastro[key].idade = idade
        if email is not None:
            sistema.cadastro[key].email = email 
        if nome is not None:
            sistema.cadastro[key].nome = nome
            # Alterando valor da chave 
            sistema.cadastro[nome] = sistema.cadastro.pop(key)
        
    def imprime_clientes(self):
        for cliente in sistema.cadastro.values():
            print(f"{cliente}\n")
        
````

## Questão 10
    Crie uma classe ContaCorrente com os atributos cliente (que deve ser um objeto da classe Cliente) e saldo. Crie métodos para depósito, saque e transferência. Os métodos de saque e transferência devem verificar se é possível realizar a transação.

````python
class ContaCorrente:
    def __init__(self, cliente, saldo=0):
        '''
        Construtor
        
        Parâmetros
        ------------
        cliente : obj
            Objeto da classe Cliente
        saldo : float
            Saldo do cliente na criação da conta
        '''
        
        self.cliente = cliente
        self.__saldo = saldo
        
    # getter para ler o valor do saldo
    @property
    def saldo(self):
        return self.__saldo

    def __repr__(self):
        return f"{self.cliente}\nSaldo: {self.__saldo}"

    def deposito(self, valor):
        '''
        Realizar depósito na conta do self
        
        Parâmetros
        -----------
        valor : float
            Valor a ser depositado na conta
        '''
        self.__saldo += valor
    
    def saque(self, valor):
        '''
        Realizar saque na conta do self
        
        Parâmetros
        -----------
        valor : float
            Valor a ser sacado da conta
        '''
        if self.__saldo - valor >= 0:
            self.__saldo -= valor
        else:
            print('Saldo insuficiente.')
    
    def transferencia(self, other, valor):
        '''
        Realiza transferência de valor entre contas
        
        Parâmetros
        -----------
        other : ojt
            Objeto da classe cliente
        valor : float
            Valor a ser transferido
        '''
        if self.__saldo - valor >= 0:
            self.__saldo -= valor
            other.__saldo += valor
        else:
            print('Saldo insuficiente.')
````

## Questão 11
    Crie uma classe Fração cujos atributos são numerador (número de cima) e denominador (número de baixo).

    Implemente os métodos de adição, subtração, multiplicação, divisão que retornam objetos do tipo Fração.

    Implemente também o método _ repr _.

    Implemente métodos para comparação: igualdade (==) e desigualdades (!=, <=, >=, < e >).

````python
class Fracao:
    def __init__(self, numerador, denominador):
        '''
        Construtor

        Parâmetros
        -----------
        numerador : float
            numerador da fração
        denominador : float
            denominador da fração
        '''

        # Denominador deve ser diferente de 0
        if denominador == 0:
            raise ValueError('Denominador precisa ser diferente de 0.')

        self.numerador = numerador
        self.denominador = denominador
    
    def __repr__(self):
        return f"{self.numerador}/{self.denominador}"
    
    def valor(self):
        '''
        Converte a fração em decimal
        '''
        return self.numerador/self.denominador
    
    def __add__(self, other):
        num = self.numerador * other.denominador +\
        self.denominador * other.numerador
        den = self.denominador * other.denominador
        
        return f"{num}/{den}"
    
    def __sub__(self, other):
        num = self.numerador * other.denominador -\
        self.denominador * other.numerador
        den = self.denominador * other.denominador
        
        return f"{num}/{den}"
    
    def __mul__(self, other):
        num = self.numerador * other.numerador
        den = self.denominador * other.denominador
        
        return f"{num}/{den}"
    
    def __truediv__(self, other):
        num = self.numerador * other.denominador
        den = self.denominador * other.numerador
        
        return f"{num}/{den}"
    
    def __eq__(self, other):
        return self.valor() == other.valor()
    
    def __ne__(self, other):
        return self.valor() != other.valor()
    
    def __lt__(self, other):
        return self.valor() < other.valor()
    
    def __le__(self, other):
        return self.valor() <= other.valor()
    
    def __gt__(self, other):
        return self.valor() > other.valor()  
    
    def __ge__(self, other):
        return self.valor() >= other.valor()
````

````python
f1 = Fracao(3, 5)
f2 = Fracao(7, 9)

print(f1 + f2)
print(f1 - f2)
print(f1 * f2)
print(f1 / f2)
print(f1 == f2)
print(f1 != f2)
print(f1 < f2)
print(f1 <= f2)
print(f1 > f2)
print(f1 >= f2)
````

## Questão 12
    Crie uma classe Data cujos atributos são dia, mês e ano. Implemente métodos _ repr _ e para comparação: igualdade (==) e desigualdades (!=, <=, >=, < e >).

````python
class Data:
    def __init__(self, dia, mes, ano):
        '''
        Construtor
        
        Parâmetros
        -----------
        dia : int
            Dia da data
        mes : int
            Mês da data
        ano : int
            Ano da data
        '''
        # Verificações de entrada dos valores
        if (not isinstance (dia, int)) or ( dia < 1 or dia > 31):  
            raise ValueError("Dia deve ser inteiro entre 1 e 31") 
        if (not isinstance (mes, int)) or (mes < 1 or mes > 12): 
            raise ValueError("Mês deve ser inteiro entre 1 e 12")  
        if (not isinstance (ano, int)) or ( ano < 0):
            raise ValueError("Ano deve ser inteiro e positivo")
        
        self.dia = dia
        self.mes = mes
        self.ano = ano
    
    def __repr__(self):
        return f"{self.dia:02d}/{self.mes:02d}/{self.ano}"
    
    def __eq__(self, other):
        if (self.ano == other.ano) and (self.mes == other.mes) \
        and (self.dia == other.dia):
            return True
        return False
    
    def __ne__(self, other):
        if (self.ano != other.ano) or (self.mes != other.mes) \
        or (self.dia != other.dia):
            return True
        return False
    
    def __lt__(self, other):
        if self.ano > other.ano:
            return False
        elif self.ano == other.ano:
            if self.mes > other.mes:
                return False
            elif self.mes == other.mes:
                if (self.dia > other.dia) or (self.dia == other.dia):
                    return False
        return True
    
    def __le__(self, other):
        if self.ano > other.ano:
            return False
        elif self.ano == other.ano:
            if self.mes > other.mes:
                return False
            elif self.mes == other.mes:
                if (self.dia > other.dia):
                    return False
        return True
    
    def __gt__(self, other):
        if self.ano < other.ano:
            return False
        elif self.ano == other.ano:
            if self.mes < other.mes:
                return False
            elif self.mes == other.mes:
                if (self.dia < other.dia) or (self.dia == other.dia):
                    return False
        return True
    
    def __ge__(self, other):
        if self.ano < other.ano:
            return False
        elif self.ano == other.ano:
            if self.mes < other.mes:
                return False
            elif self.mes == other.mes:
                if (self.dia < other.dia):
                    return False
        return True 
            
````

````python
class ContaVip(ContaCorrente):
    def __init__(self, cliente, saldo=0, cheque_especial=0):
        '''
        Contrutor
        
        Parâmetros
        -----------
        cliente : obj
            Objeto da classe Cliente
        saldo : float
            Saldo do cliente na criação da conta
        cheque_especial : float
            Valor do cheque especial
        '''
        super().__init__(cliente, saldo)
        self.cheque_especial = cheque_especial
    
    def saque(self, valor):
        '''
        Realizar saque na conta do self
        
        Parâmetros
        -----------
        valor : float
            Valor a ser sacado da conta
        '''
        if((self.saldo + self.cheque_especial) - valor >= 0):
            self.deposito(-valor)
        else:
            print('Saldo insuficiente.')
            
    def transferencia(self, other, valor):
        '''
        Realiza transferência de valor entre contas
        
        Parâmetros
        -----------
        other : ojt
            Objeto da classe cliente
        valor : float
            Valor a ser transferido
        '''        
        if((self.saldo+self.cheque_especial) - valor >= 0):
            self.deposito(-valor)
            other.deposito(valor)
        else:
            print('Saldo insuficiente.') 
````
